# Titanic

```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data upload

```python
data = pd.read_csv('Titanic-Dataset.xls')
data.info()
data.describe()
```

### Data parsing

```python
data['Survived'].unique()
```

### Data visualization

```python
plt.figure(figsize=(8, 5))
sns.histplot(data, x="Age", kde=True)

plt.figure(figsize=(8, 5))
sns.histplot(data, x="Age", kde=True, hue='Survived')
```

```python
plt.figure(figsize=(4, 3))
sns.countplot(x='Sex', data=data)
plt.figure(figsize=(8, 5))
sns.countplot(x='Embarked', data=data)
plt.figure(figsize=(8, 5))
sns.countplot(x='Age', data=data)
plt.figure(figsize=(8, 5))
sns.countplot(data, x="Sex", hue="Survived", stat="percent")
plt.show()
```
